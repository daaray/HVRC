```python
entries_per_pax = Counter(pax_entries)
pax = pd.Series(list(entries_per_pax.keys()))
entries = pd.Series(list(entries_per_pax.values()))
  
data = { 'PAX': pax, 'ENTRIES': entries }     
df = pd.DataFrame(data) 
df.plot.bar(x='PAX', y='ENTRIES', figsize = (20, 10))
```

```python
picks = {x:0 for x in pax_entries}
spaces = " " * 100

for x in range(1000):
    selection = random.choice(pax_entries)
    print(f"Testing the randomness: {selection} {spaces}", end="\r", flush=True)
    time.sleep(.005)
    picks[selection] += 1

print(f"Testing complete! {spaces}", end="\r", flush=True)

pax_series = pd.Series(list(picks.keys()))
tickets_series = pd.Series(list(picks.values()))
  
data = { 'PAX': pax_series, 'PULLS': tickets_series }     
df = pd.DataFrame(data) 
df.plot.bar(x='PAX', y='PULLS', figsize = (20, 10))
```
